```python
import requests
```

```python
cric=requests.get(url)
```

```python
soup=BeautifulSoup(cric.text,'html')
```

```python
print(soup)
```

```python
para=soup.find_all('p')
```

```python
print(para)
```

```python
paragraph_texts = [p.get_text() for p in para]
```

```python
print(paragraph_texts)
```

```python
header=soup.find_all('h1')
```

```python
print(header)
```

```python
header_texts=[h.get_text() for h in header]
```

```python
print(header_texts)
```
